```python
path = Path()
learn_inf = load_learner(path/'testing2.pkl', cpu=True)
txt_input = widgets.Text()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
pred_data = {"avg_borrowRate" : [1], "avg_amount_borrowed" : [2], "num_borrows" : [3], "num_prev_defaults" : [4]}
pred_df = pd.DataFrame(pred_data)
pred_df_true = pred_df.isnull().sum()/len(pred_df)
```

```python
cred = float(learn_inf.predict(pred_df_true)[2])
cred
```

```python
if cred >= 0.9:
    print("Amazing credit score!")
    display(AmazingCredit)
elif cred >=0.7:
    print("Good credit score!")
    display(GoodCredit)
else:
    print ("Bad credit score!")
    display(BadCredit)
```
