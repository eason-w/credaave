```python
path = Path()
learn_inf = load_learner(path/'testing.pkl', cpu=True)
txt_input = widgets.Text()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
pred_data = {"avg_borrowRate" : [1], "avg_amount_borrowed" : [2], "num_borrows" : [3], "num_prev_defaults" : [4]}
pred_df = pd.DataFrame(pred_data)
pred_df_true = pred_df.isnull().sum()/len(pred_df)
learn_inf.predict(pred_df_true)
```

```python
def on_text_change(change):
    lbl_pred.value = ''
    out_pl.clear_output()
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
```

```python
txt_input.observe(on_text_change, names=['data'])
```

```python
display(VBox([widgets.Label('Input a valid Ethereum address:'), txt_input, out_pl, lbl_pred]))
```
